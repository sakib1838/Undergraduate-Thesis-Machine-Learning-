```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C://Users//SAK//Documents//Python Scripts//thesis reh//train.csv')
```

```python
x=data.iloc[0:,0:20]
y=data.iloc[0:,-1:]
```

```python
df1=data[['ram','px_height','battery_power','px_width','mobile_wt','int_memory','sc_w','talk_time','fc','sc_h','price_range']]
```

```python
df1
```

```python
x1=df1.drop(['price_range'],axis=1)
y1=df1['price_range']
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.3,random_state=2)
```

```python
clf=RandomForestClassifier(n_estimators=100)
```

```python
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.3,random_state=2)
clf=RandomForestClassifier(n_estimators=33,max_depth=10)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
trainacc=[]
testacc=[]
count=-1;
for depth in range(1,35):
    for estimator in range(1,35):
        count+=1
        rfM=RandomForestClassifier(max_depth=depth,n_estimators=estimator,random_state=2)
        rfM.fit(X_train,y_train)
        trainacc.append(rfM.score(X_train,y_train))
        testacc.append(rfM.score(X_test,y_test))
        print('max_depth:'+str(depth)+',estimator:'+str(estimator)+',trainacc:'
              +str(trainacc[count])+',testacc:'+str(testacc[count]))
```

```python
fr=pd.DataFrame({'depth':range(1,1225),'n_estimator':range(1,1225),'train':trainacc,'test':testacc})
fr
```
