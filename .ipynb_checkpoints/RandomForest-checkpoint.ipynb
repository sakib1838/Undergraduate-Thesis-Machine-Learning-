```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('C://Users//SAK//Documents//Python Scripts//thesis reh//train.csv')
```

```python
x=data.iloc[0:,0:20]
y=data.iloc[0:,-1:]
```

```python
df1=data[['ram','px_height','battery_power','px_width','mobile_wt','int_memory','sc_w','talk_time','fc','sc_h','price_range']]
```

```python
df1
```

```python
x1=df1.drop(['price_range'],axis=1)
y1=df1['price_range']
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.3)
```

```python
clf=RandomForestClassifier(n_estimators=100)
```

```python
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix_result = confusion_matrix(y_test,y_pred)
confusion_matrix_result
```

```python
classification_report_result = classification_report(y_test,y_pred)
print(classification_report_result)
```

```python
import numpy as np
x = np.linspace(0, 500, 600)

fig = plt.figure(figsize=(40,20))
plt.plot(x, y_test)
plt.title('x-axis (Actual Values) vs y-axis (Predicted Values)',fontsize=24)
plt.xlabel('Actual',fontsize=24)
plt.ylabel('Predicted',fontsize=24)
plt.plot(x, y_pred, 'o');
```

```python
import numpy as np
x = np.linspace(0, 100, 100)

fig = plt.figure(figsize=(40,20))
plt.plot(x, y_test[0:100])
plt.title('x-axis (Actual Values) vs y-axis (Predicted Values)',fontsize=24)
plt.xlabel('Actual',fontsize=24)
plt.ylabel('Predicted',fontsize=24)
plt.plot(x, y_pred[0:100], 'o');
```
